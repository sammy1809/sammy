```python
import random
```

```python
number = random.randint(1,20)
player_name= input("Hello! What's your name?")
No_of_guesses = 0
print("Okay!",player_name,"now it's time for the game, I am guessing a number between 1 to 20.")
while No_of_guesses <5:
    guess = int(input())
    No_of_guesses+=1
    if guess < number:
        print("Your guess is too low")
    if guess > number:
        print("Your guess is too high")
    if guess == number:
       break
if guess == number:
    print("you guessed the number in" + str(No_of_guesses) + "tries!")
else:
    print("You didn't guessed the number , the number was"+ str(number))
        
```
