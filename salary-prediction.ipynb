## Concept- Supervised Machine Learning and Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
url = "https://raw.githubusercontent.com/CSCult/salary-prediction/main/Salary_Data.csv"
df = pd.read_csv(url)
```

```python

# Visualising the dataset
df.plot(x='YearsExperience', y='Salary', style='o')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python

# Data Processing
X=df.iloc[: , : -1].values
Y=df.iloc[: , 1].values
```

```python

# Model Training
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
reg= LinearRegression()
reg.fit(X_train, Y_train)
```

```python

# Plotting the line of regression
line= reg.coef_*X + reg.intercept_  # y = mx + c
plt.scatter(X, Y)
plt.plot(X, line, color='Green')
plt.show()
```

```python
# Testing Data
print(X_test)
y_pred= reg.predict(X_test)
```

```python

# Comparing actual result to the predicted model result
data_frame= pd.DataFrame({'Actual':Y_test, 'Predicted':y_pred})
data_frame
```

```python
# Estimating training training and test score
print("Training Score:", reg.score(X_train, Y_train))
print("Test Score:", reg.score(X_test, Y_test))
```

```python
# Predicting score for 7.0 YearsExperience
years= [[7.0]]
own_pred= reg.predict(years)
print("Predicted Salary for the year you entered: {}".format(own_pred[0]))
```
